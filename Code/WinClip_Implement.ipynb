```python
import os
import time
import pandas as pd
import subprocess
import re

# 定義 VisA 類別與 k-shot 設定
visa_classes = [
    "candle", "capsules", "cashew", "chewinggum",
    "fryum", "macaroni1", "macaroni2",
    "pcb1", "pcb2", "pcb3", "pcb4", "pipe_fryum"
]

k_shot_list = [0, 1, 5, 10]

# 初始化結果列表
results = []

# 開始實驗
start_time = time.time()

for k_shot in k_shot_list:
    for cls in visa_classes:
        print(f"\n====== Running WinCLIP on class: {cls}, k-shot: {k_shot} ======")
        try:
            # 執行命令並擷取 stdout
            result = subprocess.run([
                "python", "eval_WinCLIP.py",
                "--dataset", "visa",
                "--class-name", cls,
                "--k-shot", str(k_shot),
                "--img-resize", "240",
                "--img-cropsize", "240",
                "--resolution", "400",
                "--batch-size", "64",
                "--vis", "True",
                "--load-memory", "True",
                "--cal-pro", "False",
                "--experiment_indx", "0",
                "--gpu-id", "0",
                "--root-dir", "./result_winclip"
            ], capture_output=True, text=True, check=True)

            output = result.stdout

            # 用正規表示法擷取數值
            metrics = {}
            pattern = re.compile(rf"{cls}=======([a-z_]+): ([\d.]+)")
            for key, value in pattern.findall(output):
                metrics[key.strip()] = float(value)

            # 保存進表格列表
            results.append({
                "class": cls,
                "k_shot": k_shot,
                "i_roc": metrics.get("i_roc", None),
                "p_roc": metrics.get("p_roc", None),
                "p_pro": metrics.get("p_pro", None),
                "i_f1": metrics.get("i_f1", None),
                "p_f1": metrics.get("p_f1", None),
                "r_f1": metrics.get("r_f1", None)
            })

        except subprocess.CalledProcessError as e:
            print(f"❌ Error occurred while running class: {cls}, k-shot: {k_shot}")
            print(e.output)
            results.append({
                "class": cls,
                "k_shot": k_shot,
                "i_roc": None,
                "p_roc": None,
                "p_pro": None,
                "i_f1": None,
                "p_f1": None,
                "r_f1": None
            })

# 建立 pandas dataframe
df = pd.DataFrame(results)

# 顯示前幾列作確認
from IPython.display import display
display(df.head())

# 儲存成 CSV 檔
csv_path = "/content/winclip_results.csv"
df.to_csv(csv_path, index=False)
print(f"\n✅ 所有實驗完成！結果已儲存至: {csv_path}")
print("🕒 總花費時間：{:.2f} 秒".format(time.time() - start_time))
```

```python
import pandas as pd
import glob
import os
import re

# 切換到目錄
csv_dir = "/content/drive/MyDrive/Colab Notebooks/CLIP_Anomaly_ZSAD/WinClip-master/result_winclip/csv"
os.chdir(csv_dir)

# 取得所有 csv 檔案
csv_files = sorted(glob.glob("visa-k-*-indx-0.csv"))
dfs = []

# 使用正規表達式提取 k 值
pattern = re.compile(r"visa-k-(\d+)-indx-0\.csv")

for file in csv_files:
    try:
        filename = os.path.basename(file)
        match = pattern.match(filename)
        if match:
            k_shot = int(match.group(1))
            df = pd.read_csv(file)
            df["k_shot"] = k_shot
            dfs.append(df)
        else:
            print(f"⚠️ 檔名不符格式：{filename}")
    except Exception as e:
        print(f"⚠️ 無法處理檔案 {file}: {e}")

# 合併所有結果
if dfs:
    df_all = pd.concat(dfs, ignore_index=True)
    df_all.to_csv("WinClip_results_all.csv", index=False)
    print("✅ 已成功合併並儲存為 WinCLip_results_all.csv")
else:
    print("❌ 沒有任何符合的檔案被載入，請確認路徑與檔案名稱格式")
```
