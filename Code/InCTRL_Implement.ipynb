```python
import os
import time
import pandas as pd
import subprocess
import re
import matplotlib.pyplot as plt
import seaborn as sns

# === 基本設定 ===
k_shot_list = [2, 4, 8]
visa_classes = [
    "candle", "capsules", "cashew", "chewinggum", "fryum",
    "macaroni1", "macaroni2", "pcb1", "pcb2", "pcb3", "pcb4", "pipe_fryum"
]
results = []
start_time = time.time()

# === 批次執行測試 ===
for k in k_shot_list:
    for cls in visa_classes:
        print(f"\n====== Running InCTRL on class: {cls}, k-shot: {k} ======")
        few_shot_dir = f"/content/drive/MyDrive/Colab Notebooks/CLIP_Anomaly_ZSAD/trained_on_visa/{k}/{cls}/"
        pt_path = os.path.join(few_shot_dir, f"{cls}.pt")
        if not os.path.exists(pt_path):
            print(f"缺少 {pt_path}，略過此項")
            continue

        val_normal_path = f"/content/drive/MyDrive/Colab Notebooks/CLIP_Anomaly_ZSAD/datasets/AD_json/{cls}_val_normal.json"
        val_outlier_path = f"/content/drive/MyDrive/Colab Notebooks/CLIP_Anomaly_ZSAD/datasets/AD_json/{cls}_val_outlier.json"

        try:
            result = subprocess.run([
                "python", "test.py",
                "--category", cls,
                "--few_shot_dir", few_shot_dir,
                "--val_normal_json_path", val_normal_path,
                "--val_outlier_json_path", val_outlier_path
            ], capture_output=True, text=True, check=True)

            output = result.stdout

            pattern = re.compile(r"AUC-ROC: ([\d.]+), AUC-PR: ([\d.]+), F1: ([\d.]+), Precision: ([\d.]+), Recall: ([\d.]+)")
            match = pattern.search(output)
            if match:
                auc_roc, auc_pr, f1, prec, recall = map(float, match.groups())
                metrics = {
                    "i_roc": auc_roc,
                    "p_pro": auc_pr,
                    "i_f1": f1,
                    "p_roc": prec,
                    "r_f1": recall
                }
            else:
                print("找不到指標輸出")
                metrics = {}

            results.append({
                "class": cls,
                "k_shot": k,
                **metrics
            })

        except subprocess.CalledProcessError as e:
            print(f" Error: {cls}, {k} - {e}")
            results.append({
                "class": cls,
                "k_shot": k,
                "i_roc": None,
                "p_roc": None,
                "p_pro": None,
                "i_f1": None,
                "r_f1": None
            })

# === 儲存 CSV ===
df_all = pd.DataFrame(results)
df_all.to_csv("InCTRL_results_all.csv", index=False)
print("✅ 結果儲存完成")
print("🕒 耗時：{:.2f} 秒".format(time.time() - start_time))

# === 原始折線圖 ===
plt.figure(figsize=(10, 6))
for k in k_shot_list:
    subset = df_all[df_all["k_shot"] == k]
    plt.plot(subset["class"], subset["i_roc"], marker='o', label=f"shot={k}")
plt.title("AUC-ROC per category")
plt.xlabel("Category")
plt.ylabel("AUC-ROC")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("plot_aucroc_by_shot.png")

plt.figure(figsize=(10, 6))
for k in k_shot_list:
    subset = df_all[df_all["k_shot"] == k]
    plt.plot(subset["class"], subset["i_f1"], marker='o', label=f"shot={k}")
plt.title("F1 Score per category")
plt.xlabel("Category")
plt.ylabel("F1 Score")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("plot_f1_by_shot.png")

# === Seaborn 美化：平均 AUC-ROC / F1 分數柱狀圖 ===
sns.set(style="whitegrid")

# 平均 AUC-ROC
plt.figure(figsize=(8, 5))
sns.barplot(data=df_all, x="k_shot", y="i_roc", ci="sd", palette="Blues_d")
plt.title("Average AUC-ROC by Shot")
plt.ylabel("AUC-ROC")
plt.xlabel("k-shot")
plt.ylim(0.5, 1.05)
plt.tight_layout()
plt.savefig("bar_avg_aucroc.png")

# 平均 F1
plt.figure(figsize=(8, 5))
sns.barplot(data=df_all, x="k_shot", y="i_f1", ci="sd", palette="Greens_d")
plt.title("Average F1 Score by Shot")
plt.ylabel("F1 Score")
plt.xlabel("k-shot")
plt.ylim(0.5, 1.05)
plt.tight_layout()
plt.savefig("bar_avg_f1.png")
```
