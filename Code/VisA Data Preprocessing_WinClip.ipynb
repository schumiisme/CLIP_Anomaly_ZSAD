```python
import os

# 根目錄
root_dir = "/content/drive/MyDrive/Colab Notebooks/CLIP_Anomaly_ZSAD/WinClip-master/datasets/VisA_pytorch/1cls"

# 要檢查的子路徑
subfolders = [
    ("train/good", "train_good"),
    ("test/good", "test_good"),
    ("test/bad", "test_bad"),
    ("ground_truth/bad", "gt_bad")
]

# 結果統整
result = {}

for class_name in sorted(os.listdir(root_dir)):
    class_path = os.path.join(root_dir, class_name)
    if not os.path.isdir(class_path):
        continue

    result[class_name] = {}
    for subfolder, label in subfolders:
        folder_path = os.path.join(class_path, subfolder)
        if os.path.exists(folder_path) and os.path.isdir(folder_path):
            files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.png'))]
            result[class_name][label] = len(files)
        else:
            result[class_name][label] = 0

# 顯示成表格
import pandas as pd
df = pd.DataFrame(result).T.fillna(0).astype(int)
df = df[["train_good", "test_good", "test_bad", "gt_bad"]]  # 指定欄位順序
import IPython.display as display
display.display(df)
```
